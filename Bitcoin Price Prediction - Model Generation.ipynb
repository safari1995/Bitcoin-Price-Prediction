# Models
in this notebook we are going to train multiple models and investigate their accuracy.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from tqdm import tqdm
from datetime import datetime
warnings.filterwarnings("ignore")
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 200)
np.set_printoptions(suppress=True)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('Dataset.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date' , inplace=True)
df
```

# holdout splitting
in the next cell we are going to split the dataset into train, test and validation, please be informed that since we are working with time series we must avoid shuffling dataset or this model will be biased.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_percentage_error , r2_score
```

```python
y = np.asarray(df['BTC/USD']).reshape(-1, 1)
X = np.asarray(df.drop(columns=['BTC/USD']))
scaler = MinMaxScaler()
y = scaler.fit_transform(y)
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=.15 )
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, shuffle=False, test_size=.15)

print(f'Train shape: {X_train.shape} \nValidation Shape: {X_val.shape} \nTest Shape: {X_test.shape}')
```

# LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from collections import deque
from tensorflow.keras.metrics import MeanAbsolutePercentageError
```

```python
y_train.shape
```

```python
XT = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
XV = np.reshape(X_val, (X_val.shape[0], X_val.shape[1], 1))
XTE = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
XTE.shape
```

```python
model = Sequential()
model.add(LSTM(64))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss='mean_absolute_error', optimizer='Adam' ,  metrics=[MeanAbsolutePercentageError()])
```

```python
history = model.fit(XT, y_train, batch_size=10, epochs=10, validation_data=(XV, y_val) )
```

```python
y_pred_lstm =model.predict(XTE)
```

```python
print('MAPE:', mean_absolute_percentage_error(y_test,y_pred_lstm) ,'R2:', r2_score(y_test,y_pred_lstm) )
```

# SVM

```python
from sklearn.svm import SVR
```

```python
svr_model = SVR(verbose=True, C=1.0, epsilon=.001, kernel='linear' , cache_size=1000)
svr_model.fit(X_train, y_train)
svr_model.score(X_test, y_test)
```

```python
y_pred_svm = svr_model.predict(X_test)
print('MAPE:', mean_absolute_percentage_error(y_test,y_pred_svm) ,'R2:', r2_score(y_test,y_pred_svm) )
```

# Gaussian Regression GPR

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import ExpSineSquared
```

```python
kernel = ExpSineSquared(length_scale=0.1 , periodicity=100 )
gpr_model = GaussianProcessRegressor(kernel=kernel , alpha=1e-5 , optimizer='fmin_l_bfgs_b' , n_restarts_optimizer=1, normalize_y=False)
gpr_model.fit(X_train , y_train)
gpr_model.score(X_val,y_val)
```

```python
y_pred_gpr = gpr_model.predict(X_test)
print('MAPE:', mean_absolute_percentage_error(y_test,y_pred_gpr) ,'R2:', r2_score(y_test,y_pred_gpr) )
```
